```python
import time

def BubbleSort(arr):
    for i in range(len(arr)-1):
        for j in range(len(arr) -1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def SelectionSort(arr):
    for i in range(len(arr)):
        min_ind = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_ind]:
                min_ind = j
        if arr[i] != arr[min_ind]:
            arr[i], arr[min_ind] = arr[min_ind], arr[i]
    return arr   

def InsertionSort(arr): #[2, 1, 3, 5, 4]
    for i in range(1, len(arr)):
        currentElement = i
        preIndex = i - 1
        while preIndex >= 0 and arr[preIndex] > currentElement:
            arr[preIndex + 1] = arr[preIndex]
            preIndex -= 1
        arr[preIndex + 1] = currentElement
        return arr

def MergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        leftList = arr[:mid]
        rightList = arr[mid:]
        MergeSort(leftList)
        MergeSort(rightList)
        
        i = j = k = 0
        
        while i < len(leftList) and j < len(rightList):
            if leftList[i] < rightList[j]:
                arr[k] = leftList[i]
                i = i + 1
                k = k + 1
            else:
                arr[k] = rightList[j]
                j = j + 1
                k = k + 1
            
        while i < len(leftList):
            arr[k] = leftList[i]
            i = i + 1
            k = k + 1
            
        while j < len(rightList):
            arr[k] = rightList[j]
            j = j + 1
            k = k + 1
            
    return arr

def QuickSort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr)//2]
        leftList = [i for i in arr if i < pivot]
        middle = [i for i in arr if i == pivot]
        rightList = [i for i in arr if i > pivot]
        
        return QuickSort(leftList) + middle + QuickSort(rightSort)
        

def BinarySearch(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if key == arr[mid]:
            return mid
        elif key > arr[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1

def LinearSearch(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
    

def mainFunc():
    try:
        get_algorithm = input("Choose an algorithm: Bubble Sort, Insertion Sort, Selection Sort, Quick Sort, Merge Sort, Binary Search, Linear Search: ").strip()

        if 'search' in get_algorithm:
            listElements = [int(i) for i in input("Enter the list of elements separated by space: ").split()]
            listElements.sort()
            findElement = int(input("Enter the element to be searched: "))
        else:
            listElements = [int(i) for i in input("Enter the list of elements separated by space: ").split()]

        start_time = time.time()

        if get_algorithm.lower() == "bubble sort":
            sorted_array = BubbleSort(listElements)
            print(f"Bubble Sorted Array: {sorted_array}")
        elif get_algorithm.lower() == "selection sort":
            sorted_array = SelectionSort(listElements)
            print(f"Selection Sorted Array: {sorted_array}")
        elif get_algorithm.lower() == "insertion sort":
            sorted_array = InsertionSort(listElements)
            print(f"Insertion Sorted Array: {sorted_array}")
        elif get_algorithm.lower() == "merge sort":
            sorted_array = MergeSort(listElements)
            print(f"Merge Sorted Array: {sorted_array}")
        elif get_algorithm.lower() == "quick sort":
            sorted_array = QuickSort(listElements)
            print(f"Quick Sorted Array: {sorted_array}")
        elif get_algorithm.lower() == "binary search":
            sorted_array = BinarySearch(listElements, findElement)
            print(f"Element {findElement} found at {sorted_array}")
        elif get_algorithm.lower() == "linear search":
            sorted_array = LinearSearch(listElements, findElement)
            print(f"Element {findElement} found at {sorted_array}")
        else:
            print("Invalid Algorithm")

        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
            
       
    except ValueError as ve:
        print(f"Value Error: {ve}")
    except KeyError as ke:
        print(f"Key Error: {ke}")
    except Exception as e:
        print(f"An error occured: {e}")

mainFunc()
```
